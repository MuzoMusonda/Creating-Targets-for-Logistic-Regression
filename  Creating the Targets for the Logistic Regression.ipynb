```python
#Standardizing our data
#Scikit-learn is probably the most useful library for machine learning in 
#Python. The sklearn library contains a lot of efficient tools for machine
#learning and statistical modeling including classification, regression, 
#clustering and dimensionality reduction
```

```python
#Data CORRECTION
#All features are standardized including Dummies
#When we standardize, we lose the whole interpretability of a dummy
#Commenting this code as we don't need it

#Absenteeism_scaler will be used to substract the mean and divide by the 
#standard deviation variablesie(featurewise.)

#from sklearn.preprocessing import StandardScaler

#absenteeism_scaler = StandardScaler()
```

```python
#Our custom scaler will not standardize all inputs, but only the ones we choose.
#In this way, we will be able to preserve the Dummies untouched.
#Best practice will have been to standardize prior to creating Dummies

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin):
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
        
```

```python
#We must choose the columns to be scaled. To see what we've got,
#We go back to the last checkpoint before we standardized.
#Checking the last checkpoint


unscaled_inputs.columns.values
```

```python
#Creating a new variable that will contain the features we wish to scale

#columns_to_scale = ['Transportation Expense','Month Value','Day of the Week', 
        #'Distance to Work','Age', 'Daily Work Load Average', 'Body Mass Index' 
        #,'Children', 'Pet']
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education']        
        
```

```python
##List comprehension is a syntactic construct which allows us to create
#a list from the existing lists based on loops,conditions..etc

columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
```

```python
absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
#Overfitting occires when a model learns to predict the train data so well
#that it fails miserably on new data.We will split the data into train and 
#test, so we can asses he models accuracy on new data.
```

```python
#The inputs contains 175 observations along 14 input values
#The targets are a vector(a quantity possessing both magnitude and direction) of length 525.
#This contains 175 observations, 14 inputs and 1 target value/obervation

print(x_test.shape, y_test.shape)
```

```python
#Logistic regression with Sklearn.Will import for machine learning 
#to tackle the mathematical issues.

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#Training the model

reg = LogisticRegression()
```

```python
#Fit the model according to the given training data

reg.fit(x_train,y_train)
```

```python
#based in the data we used, our model learned to classify ~80%(0.78)
#of the observation correctly

reg.score(x_train,y_train)
```
